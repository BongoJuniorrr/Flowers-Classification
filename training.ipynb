```python
import os
import pickle
from PIL import Image
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
train_path = 'dataset/train'
valid_path = 'dataset/val'
classes = {
0 : "astilbe",
1 : "bellflower",
2 : "black_eyed_susan",
3 : "calendula",
4 : "california_poppy",
5 : "carnation",
6 : "common_daisy",
7 : "coreopsis",
8 : "dandelion",
9 : "iris",
10 : "rose",
11 : "sunflower",
12 : "tulip",
13 : "water_lily"}

print(classes)
```

```python
def load_img(path, label):
    images = []
    labels = []
    for filename in os.listdir(path):
        img_path = os.path.join(path, filename)
        if os.path.isfile(img_path) and img_path.endswith('.jpg'):
            img = Image.open(img_path)
            img = img.convert('RGB')
            img = np.array(img)
            images.append(img)
            labels.append(label)
    return images, labels

def load_dataset(path, classes):
    dataset_images = []
    dataset_labels = []
    for label, class_name in classes.items():
        class_path = os.path.join(path, class_name)
        if os.path.isdir(class_path):
            images, labels = load_img(class_path, label)
            dataset_images.extend(images)
            dataset_labels.extend(labels)
    return dataset_images, dataset_labels

def split_dataset(images, labels, train_ratio=0.777):
    train_images, val_images, train_labels, val_labels = train_test_split(
        images, labels, train_size=train_ratio, random_state=42, stratify=labels)
    return (train_images, train_labels), (val_images, val_labels)

#Load datasets
images, labels = load_dataset(train_path, classes)

#Split datasets
(train_images, train_labels), (val_images, val_labels) = split_dataset(images, labels)

#Save datasets using Pickle
with open('train_dataset.pkl', 'wb') as f:
    pickle.dump((train_images, train_labels), f)

with open('val_dataset.pkl', 'wb') as f:
    pickle.dump((val_images, val_labels), f)
```

```python
#Load datasets
with open('train_dataset.pkl', 'rb') as f:
    train_images, train_labels = pickle.load(f)

with open('val_dataset.pkl', 'rb') as f:
    val_images, val_labels = pickle.load(f)
```

```python
def resize_images(images):
    resized_images = []
    for image in images:
        img = Image.fromarray(image)
        img = img.resize((128, 128))
        resized_images.append(np.array(img))
    return resized_images

train_images = resize_images(train_images)
val_images = resize_images(val_images)
```

```python
for image in train_images:
    print(image.shape)
```

```python
for image in val_images:
    print(image.shape)
```

```python
#Shuffle data
from sklearn.utils import shuffle

train_images, train_labels = shuffle(train_images, train_labels, random_state=42)
```

```python
#Reduce points
train_images = np.array(train_images, dtype=np.float32) / 255.0
val_images = np.array(val_images, dtype=np.float32) / 255.0
```

```python
print(train_images[0])
```

```python
from sklearn.preprocessing import LabelBinarizer

lb = LabelBinarizer()

train_labels = lb.fit_transform(train_labels)
val_labels = lb.fit_transform(val_labels)
```

```python
import tensorflow as tf
from tensorflow.keras.models import Model
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import BatchNormalization
from tensorflow.keras.layers import AveragePooling2D
from tensorflow.keras.layers import MaxPooling2D
from tensorflow.keras.layers import Conv2D
from tensorflow.keras.layers import Activation
from tensorflow.keras.layers import Dropout
from tensorflow.keras.layers import Flatten
from tensorflow.keras.layers import Input
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import concatenate
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.optimizers import SGD
```

```python
model = Sequential()
```

```python
width = 128
height = 128
classes = 14
shape = (width, height, 3)
```

```python
model.add(Conv2D(128, (3, 3), padding="same", input_shape=shape))
model.add(Activation("relu"))
model.add(BatchNormalization())
model.add(Conv2D(128, (3, 3), padding="same"))
model.add(Activation("relu"))
model.add(BatchNormalization())
model.add(MaxPooling2D(pool_size=(2,2)))
model.add(Conv2D(256, (3, 3), padding="same"))
model.add(Activation("relu"))
model.add(BatchNormalization())
model.add(Conv2D(256, (3, 3), padding="same"))
model.add(Activation("relu"))
model.add(BatchNormalization())
model.add(MaxPooling2D(pool_size=(2,2)))
model.add(Conv2D(512, (3, 3), padding="same"))
model.add(Activation("relu"))
model.add(BatchNormalization())
model.add(Conv2D(512, (3, 3), padding="same"))
model.add(Activation("relu"))
model.add(BatchNormalization())
model.add(MaxPooling2D(pool_size=(2,2)))
model.add(Flatten())
model.add(Dense(512))
model.add(Activation("relu"))
model.add(BatchNormalization())
model.add(Dense(classes))
model.add(Activation("softmax"))
```

```python
model.summary()
```

```python
gen = ImageDataGenerator(rotation_range=0.5, zoom_range=0.5, width_shift_range=0.5, height_shift_range=0.5, horizontal_flip=True)
```

```python
learning_rate = 0.01
epochs = 20
batch_size = 64
optimize = SGD(learning_rate=learning_rate, momentum=0.9)
```

```python
model.compile(optimizer=optimize, loss="categorical_crossentropy", metrics=["accuracy"])
```

```python
history = model.fit(
    x=train_images,  # Training data
    y=train_labels,  # Training labels
    batch_size=batch_size,
    epochs=epochs,
    verbose=1,
    validation_data=(val_images, val_labels)  # Validation data
)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.plot(history.history['accuracy'], label='Train Accuracy')
plt.plot(history.history['val_accuracy'], label='Val Accuracy')
plt.title('Training and Validation Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(history.history['loss'], label='Train Loss')
plt.plot(history.history['val_loss'], label='Val Loss')
plt.title('Training and Validation Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()


model.save('Flower_Classification_Large_3.h5')

# Show plots
plt.tight_layout()
plt.show()
```
